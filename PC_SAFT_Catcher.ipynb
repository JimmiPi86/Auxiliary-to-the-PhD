# Upload SAFT result

```python
import pandas as pd
import numpy as np
import os
file_path = 'MergedData.xlsx'
sheet_nameR = 'Rho'
sheet_nameU = 'U'
dfr = pd.read_excel(file_path, sheet_name=sheet_nameR)
dfu = pd.read_excel(file_path, sheet_name=sheet_nameU)
import os
```

```python
name = '2Hea_Pr'
```

```python
dfr
```

```python
dfr['p'] = dfr['p']/10
dfu['p'] = dfu['p']/10
```

```python
dfu
```

# Załaduj dane, odnoszące się do eksperymantalnych ciśnień

```python
#Frame Geśtości
PR = [0.1,10,15,20,25,30,35]
#Frame predkości
PU  = [0.10,1.00,2.00,3.00,4.00,5.00,6.00,7.00,8.00,9.00,10.00,11.00,12.00,13.00,14.00,15.00,16.00,17.00,18.00,19.00,20.00]
```

```python
resr = dfr[dfr['p'].isin(PR)]
resu = dfu[dfu['p'].isin(PU)]
```

```python
resr
```

# Spłaszczenie tabel po izobarach


```python
resr_without_p = resr.drop('p', axis=1)
resu_without_p = resu.drop('p', axis=1)
resr_flat = np.array(resr_without_p).flatten()  # przekształcenie do jednowymiarowej tablicy numpy
resu_flat = np.array(resu_without_p).flatten()
resr_numerical = [val.item() for val in resr_flat]  # wyodrębnienie wartości liczbowych
resu_numerical = [val.item() for val in resu_flat]
raw_r = pd.DataFrame(resr_numerical) ###Zapis wektora danych w postaci kolumny surowych danych
raw_u = pd.DataFrame(resu_numerical)
```

# Zapis tabeli

```python
# Nazwa katalogu
directory_name = "DATA"

# Utwórz katalog, jeśli nie istnieje
if not os.path.exists(directory_name):
    os.makedirs(directory_name)
### GĘstość
file_path = os.path.join(directory_name, "PCSAFTR_" + name + ".xlsx")
resr.to_excel(file_path, index=False)  # Ustawienie index=False usuwa indeksy w pliku Excel
#### Prędkość
file_path = os.path.join(directory_name, "PCSAFTU_" + name + ".xlsx")
resu.to_excel(file_path, index=False)  # Ustawienie index=False usuwa indeksy w pliku Excel
print(f"Plik zapisany w: {file_path}")
```

# Zapis RAW

```python
# Nazwa katalogu
directory_name = "RAW"

# Utwórz katalog, jeśli nie istnieje
if not os.path.exists(directory_name):
    os.makedirs(directory_name)
# Pełna ścieżka do pliku
file_pathr = os.path.join(directory_name, "RAWR_" + name + ".xlsx")
raw_r.to_excel(file_pathr, index=False)  # Ustawienie index=False usuwa indeksy w pliku Excel

file_pathu = os.path.join(directory_name, "RAWU_" + name + ".xlsx")
raw_u.to_excel(file_pathu, index=False)  # Ustawienie index=False usuwa indeksy w pliku Excel

print(f"Plik zapisany w: {file_path}")
```
